```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
model = LinearRegression()

# Generate initial random data points
np.random.seed(0)
X_initial = np.random.rand(10, 1) * 10
y_initial = 2 * X_initial + 1 + np.random.randn(10, 1)

# Fit the initial data points to the model
model.fit(X_initial, y_initial)

# Predictions based on the initial data
y_pred_initial = model.predict(X_initial)

# Create a scatter plot of the initial data points
plt.scatter(X_initial, y_initial, label="Data Points", color='blue')

# Plot the initial regression line
plt.plot(X_initial, y_pred_initial, label="Initial Regression Line", color='green')

plt.xlabel("X")
plt.ylabel("y")
plt.title("Interactive Linear Regression")
plt.legend()
plt.show()


```

```python
# Interactive loop to add new data points
plt.ion()  # Turn on interactive mode

while True:
    new_x = input("Enter a new x value (or 'exit' to quit): ")
    if new_x.lower() == 'exit':
        break

    try:
        new_x = float(new_x)
        new_y = float(input("Enter the corresponding y value: "))

        # Add the new data point to the existing data
        X_initial = np.append(X_initial, [[new_x]], axis=0)
        y_initial = np.append(y_initial, [[new_y]])

        # Update the model with the new data point
        model.fit(X_initial, y_initial)

        # Predictions based on the updated data
        y_pred_initial = model.predict(X_initial)

        # Clear the plot and replot everything
        plt.clf()
        plt.scatter(X_initial, y_initial, label="Data Points", color='blue')
        plt.plot(X_initial, y_pred_initial, label="Updated Regression Line", color='green')
        plt.scatter(new_x, new_y, label="New Data Point", color='red', marker='o')
        plt.xlabel("X")
        plt.ylabel("y")
        plt.title("Interactive Linear Regression")
        plt.legend()
        plt.draw()
        plt.pause(0.2)

    except ValueError:
        print("Invalid input. Please enter numeric values for x and y.")
        continue

plt.ioff()  # Turn off interactive mode
plt.show()
```
